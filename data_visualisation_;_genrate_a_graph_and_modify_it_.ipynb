```python
import numpy as np
```

```python
languages =['python' , 'C/C++', 'Javascript', 'C#']
students = [12,4,3,2]
pos = np.arange(len(students))
```

```python
import matplotlib.pyplot as plt

def basic_plot ():
  barlist= plt.bar(languages,students,)
  plt.title('language popularity in B106 WS2024')
  plt.ylabel('number of students')

  return barlist

```

```python
def remove_tricks_and_labels():
  plt.tick_params(axis ='x', bottom = False ) # set parameter
  plt.tick_params(axis ='y', left = False , labelleft = False  )
```

```python
#[s for a in  plt.gca().spines.values()]

def removes_spines():
  for spine in plt.gca().spines.values(): # give me the list of spines in this object
    spine.set_visible(False)
```

```python
def recolor_bars(barlsist, highlight_index= 0, highlight_color='red' ):
  for bar in barlist :
    bar.set_color('gray')

  barlist[highlight_index].set_color(highlight_color) # this is by default

```

```python
def recolor_labels(default_color= 'gray', highlight_index = 0, highlight_color='red' ):
  for label in plt.gca().get_xticklabels() :
    label.set_color(default_color)

  plt.gca().get_xticklabels()[highlight_index].set_color(highlight_color) # this is by default
  plt.gca().get_yaxis().get_label().set_alpha(.5)# alpha is the transparenty

def add_bar_labels(barlist):
  for i, bar in enumerate(barlist) : # get the index from the list
    x= bar.get_x() + bar.get_width()/2
    y= bar.get_y() + bar.get_height () - 1

    plt.gca().text(x,y,students[i], color= 'white' , ha = 'center', size= 12)

    #print (bar.get_height())
```

```python

plt.figure()
barlist = basic_plot()

remove_tricks_and_labels()
removes_spines()

recolor_bars(barlist , 1, 'navy')
recolor_labels('gray', 1,'navy')
add_bar_labels(barlist)

```
